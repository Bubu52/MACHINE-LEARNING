## 6. Usamos el modelo

* Volvemos a cargar el estandarizador y el modelo simulando que lo recuperamos
* Creamos los nuevos datos sobre los que queremos usar / hacer predicciones
* Los estandarizamos con el estandarizador recuperado
* Predecimos la variable 'Purchased' para esos nuevos datos

```python
# Recuperamos los codificadores, el estandarizador y el modelo
# Tu código aquí debajo
est=load('estandarizador.std')
ord=load('ordinal.trf')
hot=load('onehot.trf')
model=load('modelo_LR.md')
```

```python
# Tomamos los nuevos datos "diabetes_new_data.xlsx"
# Tu código aquí debajo
nuevos=pd.read_excel('diabetes_new_data.xlsx')
nuevos
```

```python
# Aplicamos las codificaciones a nuestros nuevos datos

# Primero la transformación de ordinales a numéricas (ordinal encoder)
# Tu código aquí debajo
nuevos['Satisfaction']=ord.transform(nuevos[['Satisfaction']])
nuevos
```

```python
# Ahora la transformación de categóricas sin orden a numéricas (one hot encoding)
# Tu código aquí debajo
a=hot.transform(nuevos[['Job']])
# Convierto el a en data frame y lo llamo encoded_df
encoded_df = pd.DataFrame(a.toarray(), columns=hot.get_feature_names_out(['Job']))

# Reseteo el índice de los dos data frames antes de concatenarlos
nuevos.reset_index(drop=True, inplace=True)
encoded_df.reset_index(drop=True, inplace=True)

# concateno los dos data frames y los guardo machacando el df
nuevos=pd.concat([nuevos,encoded_df],axis=1)

# Elimino la columna Job original antigua, la categórica
nuevos.drop(columns='Job',inplace=True)

# Muestro tamaño yu sample del data frame df transformado
print(nuevos.shape)
nuevos
```

```python
# Estandarizamos los nuevos datos
# Tu código aquí debajo

nuevos_std=est.transform(nuevos)
```

```python
# Predecimos la variable 'Outcome' para esos valores y los mostramos.
# Tu código aquí debajo

nuevos['Predicción']=model.predict(nuevos_std)
nuevos
nuevos.to_excel('Resultado_modelo.xlsx')
```
