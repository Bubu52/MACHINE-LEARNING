#### - RECOLECIÓN DE DATOS CON URLLIB-3 y CSV

```python
# Importamos libreria
import csv
import urllib3

# Descargamos el dataset y almacenamiento
csv_name = "titanic.csv"

# Request de la URL que contiene los datos
response = urllib3.request("get","https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")

# Verificamos si el "response" es de tipo 200 (tenemos datos OK)
if response.status == 200:
    print("[+] Respuesta Correcta")

    # Decodificamos los datos de bytes a string en codificacion UTF-8
    data = response.data.decode("utf-8")

    # Apertura / Creación de archivo CSV en el que almacenamos los datos
    with open(csv_name,"w", newline = "\n", encoding = "utf-8") as csv_file:

        # Definimos el objeto "writer"
        csv_writer = csv.writer(csv_file)

        #Procesamos cada linea del csv para añadirlo al archivo abierto
        for lines in data.split("\n"):
            csv_writer.writerow(lines.split(","))
    print(f"[+] Archivo cargado correctamente: {csv_name}")
else:
    print(f"Respuesta incorrecta, codigo de respuesta: {response.status}")
```
